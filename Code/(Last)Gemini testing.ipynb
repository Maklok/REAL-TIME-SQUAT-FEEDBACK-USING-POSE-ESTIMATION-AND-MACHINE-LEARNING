```python
from google.colab import userdata
from google import genai

api_key = userdata.get("GOOGLE")
client = genai.Client(api_key=api_key)

```

```python
def rephrase_feedback_with_gemini(template_text):
    prompt = (
        "Rephrase the following fitness coaching instruction into a natural, "
        "encouraging sentence. Do not change the meaning or the corrective advice.\n\n"
        f'Instruction: "{template_text}"'
    )

    response = client.models.generate_content(
        model="gemini-3-flash-preview",
        contents=prompt
    )

    return response.text.strip()
```

```python
STYLE_FEEDBACK_TEMPLATES = {
    "Back": "You are performing a back squat.",
    "Front": "You are performing a front squat.",
    "Goblet": "You are performing a goblet squat."
}


POSTURE_FEEDBACK_TEMPLATES = {
    "Correct": "Your squat technique is correct.",
    "Insufficient_depth": "You are not reaching sufficient squat depth.",
    "Rounded_back": "Your lower back is rounding. Keep a neutral spine.",
    "Weight_on_toes": "Your weight is shifting toward your toes. Keep pressure on your heels."
}
```

```python
def build_feedback_template(style_label, posture_label):
    style_text = STYLE_FEEDBACK_TEMPLATES[style_label]
    posture_text = POSTURE_FEEDBACK_TEMPLATES[posture_label]

    combined = f"{style_text} {posture_text}"
    return combined
```

```python
def rephrase_feedback_with_gemini(template_text):
    prompt = (
        "Rephrase the following fitness coaching message into a natural, clear, "
        "and encouraging sentence. Do not add new advice and do not remove any "
        "instructional content.\n\n"
        f'Message: "{template_text}"'
    )

    response = client.models.generate_content(
        model="gemini-3-flash-preview",
        contents=prompt
    )

    return response.text.strip()

```

```python
predicted_style = "Front"
predicted_posture = "Rounded_back"

template = build_feedback_template(predicted_style, predicted_posture)
final_feedback = rephrase_feedback_with_gemini(template)

print("Template:")
print(template)

print("\nGemini output:")
print(final_feedback)

```
