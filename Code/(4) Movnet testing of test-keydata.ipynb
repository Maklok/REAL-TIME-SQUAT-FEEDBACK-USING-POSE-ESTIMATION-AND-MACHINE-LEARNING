```python
import os
import random
import numpy as np
import matplotlib.pyplot as plt

DATA_DIR = "....."

files = [f for f in os.listdir(DATA_DIR) if f.endswith(".npy")]
samples = random.sample(files, 2)
```

```python
JOINT_IDX = 11  # left hip

for file in samples:
    path = os.path.join(DATA_DIR, file)
    data = np.load(path)  # (50, 17, 2)

    x = data[:, JOINT_IDX, 0]
    y = data[:, JOINT_IDX, 1]

    plt.figure()
    plt.plot(x, label="x")
    plt.plot(y, label="y")
    plt.title(f"Joint trajectory over time\n{file}")
    plt.xlabel("Frame")
    plt.ylabel("Coordinate value")
    plt.legend()
    plt.show()
```

```python
velocity = np.diff(data, axis=0)
mean_motion = np.mean(np.linalg.norm(velocity, axis=2))
print(mean_motion)
```

```python
print(data.min(), data.max())
```
