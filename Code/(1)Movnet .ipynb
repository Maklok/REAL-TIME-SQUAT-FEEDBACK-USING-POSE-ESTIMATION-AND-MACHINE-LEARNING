```python
READY_ROOT = "..."
KEYPOINT_ROOT = "..."

for root, _, files in os.walk(READY_ROOT):
    for file in files:
        if not file.lower().endswith(".mp4"):
            continue

        video_path = os.path.join(root, file)

        # Speil sti
        rel_path = os.path.relpath(video_path, READY_ROOT)
        out_path = os.path.join(
            KEYPOINT_ROOT,
            rel_path.replace(".mp4", ".npy")
        )

        os.makedirs(os.path.dirname(out_path), exist_ok=True)

        try:
            sequence = video_to_sequence(video_path)
            np.save(out_path, sequence)
            print(f"Lagret: {out_path}")
        except Exception as e:
            print(f"Feil i {video_path}: {e}")

```
